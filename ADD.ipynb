```python
import pandas as pd
original_data=pd.read_excel(r'F:/研究生阶段/毕业设计/RWTCd.xls',sheet_name='s').set_index('Date')
true_data=pd.read_excel(r'F:\研究生阶段\毕业设计\处理数据\EMD分解.xls').set_index('Date')
pre_data=pd.read_excel(r'F:\研究生阶段\毕业设计\处理数据\pred_imfs.xls').set_index('Date')
```

```python
orig_data=original_data['2008-4-1':'2013-10-16']

true_data=true_data['2008-4-1':'2013-10-16']
pre_data=pre_data['2008-4-1':'2013-10-16']
```

```python
pre_data['col_sum']=pre_data.apply(lambda x: x.sum(), axis=1)
print(pre_data['col_sum'].tail())
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(orig_data, pre_data['col_sum'])**0.5
print("RMSE", error)
```

```python
import matplotlib.pyplot as plt  
plt.figure(figsize=(18,5))
plt.title('Actual and predicted')
plt.plot(orig_data.index,orig_data, label = 'Original data')
plt.plot(orig_data.index,pre_data['col_sum'], label = 'Predicted data')
plt.legend(['Actual', 'Prediction'],loc='best')
```
